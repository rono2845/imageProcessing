```python
#Write a function, s = scanLine(f, I, loc), where f is a grayscale image, I is an integer, and loc is a character string with value ‘row’ and ‘col’ indicating that I is either a row or column number in image f. The output, s, is a vector containing the pixel values along the specified row or column in f.


def scanLine(f, I=0, loc="row"):
    if loc=="row":
        return f[I,:]
    elif loc == "col":
        return f[:,I]
    
doggo_gray = rgb2gray(doggo)
s=scanLine(doggo_gray, 9, "row")
print(s)
```

```python
#Scale the intensities of the image to the full [0, K] range

#fm = f − min(f); fs = K[ fm/max(fm)]

def contrastStretch(img, K):
    fmin = img.min()
    fmax = img.max()
    return np.array(list(map(lambda x : K * ((x - fmin)/(fmax - fmin)), img  )))
```

```python
#Display the original image, the resulting images using K = 1, and K = 255 in (2). Is there any difference among these 3 images?
#wom1_gray=rgb2gray(wom1)
K1 = contrastStretch(wom1, 1)
#plt.gray()
#plt.imshow(K1)
```

```python
'''Intensity transformation of grayscale images.

1a) Write a function (g, map) = intXform(f, mode, param) for transforming the intensities of an input 8-bit grayscale image f. The intensities of f (and output image g) are assumed to be in the range (0,1). The type of transformation performed is specified in parameter mode, a character string with values: ‘negative’, ‘log’, or ‘gamma’. These specifications implement the intensity transformations given in the lecture slides, i.e., equations (1), (2), and (3) with c = 1.0, and γ in equation (3) is specified by the argument param. On the output, g is the output image, and map is the transformation function computed by intXform. (Note the need of conversion between dynamic ranges (0,L−1) and (0,1).)
'''

def intXform(f, mode, param = 1.0):
    c = 1.0
    # Normalize input
    #norm_input = minmax_scale(f)

    # Define the different transformations
    xform = lambda x : x
    if mode == 'negative':
        xform = lambda r : -r + 255
    if mode == 'gamma':
        xform = lambda r : c * (pow(r,param))
    if mode == 'log':    
        xform = lambda r : c * np.log(r + 1.0)
        
    # Apply the transformation on each pixel
    g = np.array(list(map(xform, f)))

    return g, xform
```

```python
def cityblock_distance(A, B):
  
    result = np.sum([abs(a - b) for (a, b) in zip(A, B)])
    return result
```

```python
array1 = [1, 2, 13, 5]
array2 = [1, 27, 3, 4]
  
result = cityblock_distance(array1, array2)
  
print("The CityBlock distance between 2 arrays is:", result)
```

```python
def euclidean_distance(A, B):
    result1 = np.sum([pow((a - b),2) for (a, b) in zip(A, B)])
    result = math.sqrt(result1)
    return result
```

```python
array1 = [1, 2, 13, 5]
array2 = [1, 27, 3, 4]
  
result = euclidean_distance(array1, array2)
  
print("The euclidean distance between 2 arrays is:", result)
```

```python
def chessboard_distance(A, B):
    result = max([abs(a - b) for (a, b) in zip(A, B)])
    return result
```

```python
array1 = [1, 2, 13, 5]
array2 = [1, 27, 3, 4]
  
result = chessboard_distance(array1, array2)
  
print("The chessboard distance between 2 arrays is:", result)
```
